```python
#import the libraries

import numpy as np
import pandas as pd
from matplotlib import pyplot as plt

#Task 1
#Read the bmi.csv

bmi_df=pd.read_csv('bmi.csv')
bmi_df
```

```python
#Task 2
#Data is sorted in descending order in accordance with BMI value
#Find the top 5 age group where the BMI value is the highest, and plot a bar graph out of it

bmi_top5=bmi_df.head(5)
print(bmi_top5)

name=bmi_top5["Age"]
number=bmi_top5["BMI"]

plt.xlabel("Age")
plt.ylabel("BMI")
plt.xticks(rotation="vertical")

label=name
value=number
plt.bar(label, value, width=0.4, color=('red', 'green', 'blue', 'yellow', 'orange'))
```
